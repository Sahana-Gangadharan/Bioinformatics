```python
def string_search(string,DNA):
    m=len(string)
    n=len(DNA)
    ctr=0
    position=[]
    for i in range(n):
        if DNA[i:i+m]==string:
            ctr+=1
            position.append(i)
    print('Position of occurance = ')
    print(position)
    print('Total number of occurance = %d'%ctr)
DNA='GCATTAAGGTGTGGAGTTACACAGTTACCAGGTTA'
string='TTA'
string_search(string,DNA)
```

```python
def energy(string):
    BSE={'AA':-4,'AT':-7,'AC':-5,'AG':-11,
        'TA':-7,'TT':-2,'TC':-3,'TG':-4,
         'CA':-9,'CT':-5,'CC':-6,'CG':-7,
         'GA':-9,'GT':-6,'GC':-4,'GG':-11}
    energy=0
    n = len(string)
    for i in range(n-1):
        base=string[i:i+2]
#         print(base)
        energy+=BSE[base]
    print(energy)
    avg_energy=energy/(n-1)
    print('Average Base Stacking Energy = %f'%avg_energy)
DNA = 'GCAAGGTGTGGAGTTACA'
energy(DNA)
```

```python
def penta_match(string1, string2):
    n = len(string1)
    m = len(string2)
    occ1 = [0]*(n-4)
    occ2 = [0]*(m-4)
    for i in range(n-4):
        penta = string1[i:i+5]
        for j in range(m-4):
            if string2[j:j+5]==penta:
                occ2[i]+=1
        for k in range(n-4):
            if string1[k:k+5]==penta:
                occ1[i]+=1
    
    for k in range(len(occ1)):
        if occ1[k]>=1 and occ2[k]>=1:
            print('The number of occurance of %s in sequence 1 = %d and sequence 2 = %d.' %(string1[k:k+5],occ1[k],occ2[k]))
    
h = 'MVHLTPEEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFESFGDLSTPDAVMGNPKVKAHGKKVLGAFSDGLAHLDNLKGTFATLSELHCDKLHVDPENFRLLGNVLVCVLAHHFGKEFTPPVQAAYQKVVAGVANALAHKYH'
c = 'MVHWTAEEKQLITGLWGKVNVAECGAEALARLLIVYPWTQRFFASFGNLSSPTAILGNPMVRAHGKKVLTSFGDAVKNLDNIKNTFSQLSELHCDKLHVDPENFRLLGDILIIVLAAHFSKDFTPECQAAWQKLVRVVAHALARKYH'
penta_match(h,c)           
```

```python
def blast(str1,str2):
    similarity_AA = {'A':['C','S','T','G','V','L','I', 'M'], 'G':['A','N','S','V','L','I', 'M'], 'V':['G','A','L','I','T','M'], 
                     'L':['G','V','A','I','M','F'], 'I':['G','F','V','L','A','M'], 'M':['G','V','L','I','Q','F','A'], 
                     'S':['A','N','D','E','G','K','T','P','N','Q','C'], 'T':['A','S','P','N','V','Q','C'], 'P':['T','S','N','Q','C'],
                     'N':['R','E','H','K','T','P','S','Q','C'], 'Q':['R','D','E','H','K','M','T','P','N','S','C'], 
                     'C':['T','P','N','Q','S'], 'K':['R','N','Q','E','S','H'], 'R':['N','Q','E','K','H'], 
                     'H':['R','N','Q','E','Y','K'], 'D':['E','N','Q','S'], 'E':['R','N','D','Q','H','K','S',], 
                     'F':['I','L','M','Y','W'], 'Y':['H','F','W'], 'W':['Y','F'] }

    identity=0
    similarity=0
    i_s = []
    for i in range(max(len(str1),len(str2))):
        if str1[i]==str2[i]:
            identity+=1
            similarity+=1
            i_s.append(i)
        if str1[i] in similarity_AA[str2[i]] or str2[i] in similarity_AA[str1[i]]:
            similarity+=1
            i_s.append(i)
        print(str1[i],str2[i],similarity)
            
    gap=max(len(str1),len(str2))-min(len(str1),len(str2))
    query_cov = max(len(str1),len(str2)) - gap
    
    print('Sequence identity = %d / %d' %(identity,max(len(str1),len(str2))))
    print('Sequence similarity/positives = %d / %d' %(similarity,max(len(str1),len(str2))))
    print('Query coverage = %d / %d' %(query_cov,max(len(str1),len(str2))))
    print('Gap percentage = %d / %d' %(gap,max(len(str1),len(str2))))
    
h = 'MVHLTPEEKSAVTALWGKVNVDEVGGEALGRLLVVYPWTQRFFESFGDLSTPDAVMGNPKVKAHGKKVLGAFSDGLAHLDNLKGTFATLSELHCDKLHVDPENFRLLGNVLVCVLAHHFGKEFTPPVQAAYQKVVAGVANALAHKYH'
c = 'MVHWTAEEKQLITGLWGKVNVAECGAEALARLLIVYPWTQRFFASFGNLSSPTAILGNPMVRAHGKKVLTSFGDAVKNLDNIKNTFSQLSELHCDKLHVDPENFRLLGDILIIVLAAHFSKDFTPECQAAWQKLVRVVAHALARKYH'
# blast(h,c)

s1 = 'DERTGHFFLKSAVLIPCVIDYMNS'
s2 = 'EQKFGPYHVIEINFVASALLRVDT'
blast(s1,s2)
```

```python
import numpy as np

def pair_pref(seq):
    AA_all =['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']
    n = len(seq)
    pref1 = [[0 for i in range(20)] for i in range(20)]
    pref2 = [[0 for i in range(20)] for i in range(20)]
    pref3 = [[0 for i in range(20)] for i in range(20)]

    comp = [0]*20
    for i in range(n):
        aa = seq[i]
        ind = AA_all.index(aa)
        comp[ind]+=1

    for i in range(20):
        for j in range(20):
            pair = AA_all[i]+AA_all[j]
            for k in range(n-1):
                if pair==seq[k:k+2]:
                    pref1[i][j]+=1
                    pref2[i][j]+=1
                    pref3[i][j]+=1
            if comp[i]==0 or comp[j]==0:
                pref1[i][j]=0
                pref2[i][j]=0
                pref3[i][j]=0
            else:
                pref1[i][j]=round((pref1[i][j]*100)/(comp[i]+comp[j]),ndigits=1)
                pref2[i][j]=round((pref2[i][j]*100)/(n-1),ndigits=1)
                pref3[i][j]=round((pref3[i][j]*100)/(comp[i]*comp[j]),ndigits=1)
    
#     print('\nPair-wise preference score for the given sequnce-')
#     print('\nFormula 1 : ')
    
#     for i in range(20):
#         for j in range(20):
#             print(pref1[i][j],end=' ')
#         print()
       
#     print('\nFormula 2 : ')    
#     for i in range(20):
#         for j in range(20):
#             print(pref2[i][j],end=' ')
#         print()
    
#     print('\nFormula 3 : ')    
#     for i in range(20):
#         for j in range(20):
#             print(pref3[i][j],end=' ')
#         print()
        
    return pref1,pref2,pref3

seq1 = 'RATPTRWPVGCFNRPWTKWSYDEALDGIKAAGYAWTGLLTASKPSLHHATATPEYLAALKQKSRHAA'
seq2 = 'AAAVMMGLAAIGAAIGIGILGGKFLEGAARQPDLIPLLRTQFFIVMGLVDAIPMIAVGLGLYVMFAVA'
seq3 = 'QSGMTYRLGLSWDWDKSWWQTSTGRLTGYWDAGYTYWEGGDEGAGKHSLSFAPVFVYEFAGDSIKPFIEAGIGVAAFSGTRVGDQNLGSSLNFEDRIGAGLKFANGQSVGVRAIHYSNAGLKQPNDGIESYSLFYKIPI'

pair_pref(seq1)
```

```python
def matchingString(x,y):
    match=''
    for i in range(0,len(x)):
        for j in range(0,len(y)):
            k=1
            # now applying while condition untill we find a substring match and length of substring is less than length of x and y
            while (i+k <= len(x) and j+k <= len(y) and x[i:i+k]==y[j:j+k]):
                if len(match) <= len(x[i:i+k]):
                    match = x[i:i+k]
                k=k+1
    return match  
```
